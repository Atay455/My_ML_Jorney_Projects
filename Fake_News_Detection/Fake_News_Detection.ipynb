```python
import numpy as np
import pandas as pd
import re
import string
from sklearn.metrics import classification_report
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
import nltk
nltk.download("stopwords")
```

```python
# Printing stopwords in english
print(stopwords.words("english"))
```

**Data Pre-processing**

```python
# loading the dataset to our pandas DataFrame
news_dataset_fake = pd.read_csv("Fake.csv")
news_dataset_true = pd.read_csv("True.csv")
```

```python
news_dataset_fake["class"] = 0
news_dataset_true["class"] = 1
```

```python
news_dataset_fake.head(10)
```

```python
news_dataset_true.head(10)
```

```python
news_dataset_fake.shape, news_dataset_true.shape
```

```python
df_fake_manual_testing = news_dataset_fake.tail(10)
df_true_manual_testing = news_dataset_true.tail(10)
```

```python
news_dataset_fake = news_dataset_fake.iloc[:-10]
news_dataset_true = news_dataset_true.iloc[:-10]
```

```python
data_manual_testing = pd.concat([df_fake_manual_testing, df_true_manual_testing], axis =0)
data_merge = pd.concat([news_dataset_fake,news_dataset_true],axis=0)
```

```python
data_manual_testing.to_csv("manual_testing.csv")
```

```python
data_merge.head(10)
```

```python
data_merge = data_merge.drop(["title", "subject", "date"], axis=1)
data_merge.head(10)
```

```python
data_merge = data_merge.sample(frac=1)
```

```python
data_merge.head(10)
```

```python
data_merge.isnull().sum()
```

```python
def word_drop(text):
  text = text.lower()
  text = re.sub(r'\[.*?\]','',text)
  text = re.sub(r"\\W"," ",text)
  text = re.sub(r'https?://\S+|www\/S+','',text)
  text = re.sub(r'<.*?>+','',text)
  text = re.sub(r'[%s]'% re.escape(string.punctuation),'',text)
  text = re.sub(r'\n','',text)
  text = re.sub(r'\w*\d\w*','',text)
  return text
```

```python
data_merge["text"] = data_merge["text"].apply(word_drop)
```

```python
data_merge.head(10)
```

```python
x = data_merge["text"]
y = data_merge["class"]
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y ,test_size = 0.25)
```

```python
vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

### Logistic Regression

```python
LR = LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
LR.score(xv_test,y_test)
```

```python
predict_LR = LR.predict(xv_test)
```

```python
print(classification_report(y_test,predict_LR))
```

### Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT = DecisionTreeClassifier()
DT.fit(xv_train,y_train)
```

```python
DT.score(xv_test,y_test)
```

```python
pred_DT = DT.predict(xv_test)
```

```python
print(classification_report(y_test,pred_DT))
```

### Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
GBC = GradientBoostingClassifier(random_state=0)
GBC.fit(xv_train,y_train)
```

```python
GBC.score(xv_test,y_test)
```

```python
pred_GBC = GBC.predict(xv_test)
```

```python
print(classification_report)
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RFC = RandomForestClassifier(random_state=0)
RFC.fit(xv_train,y_train)
```

```python
RFC.score(xv_test,y_test)
```

```python
pred_RFC = RFC.predict(xv_test)
```

```python
print(classification_report(y_test,pred_RFC))
```

**Manual Testing**

```python
def output_label(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"


def manual_testing(news):
    # Put the news text into a dictionary
    testing_news = {"text": [news]}

    # Convert dictionary to DataFrame
    new_def_test = pd.DataFrame(testing_news)

    # Apply text preprocessing
    new_def_test["text"] = new_def_test["text"].apply(word_drop)

    # Extract text column
    new_x_test = new_def_test["text"]

    # Convert text into numerical vectors
    new_xv_test = vectorization.transform(new_x_test)

    # Predictions from different models
    pred_LR = LR.predict(new_xv_test)
    pred_DT = DT.predict(new_xv_test)
    pred_GBC = GBC.predict(new_xv_test)
    pred_RFC = RFC.predict(new_xv_test)

    # Print predictions
    print(
        "\n\nLR Prediction: {} \nDT Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(
            output_label(pred_LR[0]),
            output_label(pred_DT[0]),
            output_label(pred_GBC[0]),
            output_label(pred_RFC[0])
        )
    )
```

```python
news = str(input())
manual_testing(news)
```
